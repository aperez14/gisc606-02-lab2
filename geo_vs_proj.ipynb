```python
import os
from glob import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import geopandas as gpd
from shapely.geometry import Point
import earthpy as et
import seaborn as sns

sns.set(font_scale = 1.5)
sns.set_style("white")

data = et.data.get_data("spatial-vector-lidar")
os.chdir(os.path.join(et.io.HOME, "earth-analytics"))
```

```python
boulder_df = np.array([[476911.31, 4429455.35]])
geometry = [Point(xy) for xy in boulder_df]

boulder_loc = gpd.GeoDataFrame(geometry, columns=["geometry"], crs={"init": "epsg:2957"})
```

```python
fig, ax = plt.subplots(figsize=(10,10))
boulder_loc.plot(ax=ax)
ax.set_title("Plot of Boulder, CO location")
plt.ticklabel_format(useOffset=False)
plt.show()
```

```python
boulder_loc.crs
```

```python
boulder_df_geog = boulder_loc.to_crs(epsg=4326)
boulder_df_geog.crs
```

```python
worldBound_path = os.path.join("data", "spatial-vector-lidar", "global", "ne_110m_land", "ne_110m_land.shp")

worldBound = gpd.read_file(worldBound_path)
```

```python
fig, ax = plt.subplots(figsize=(10,5))

worldBound.plot(cmap="Greys", ax=ax)

ax.set(title="Bould CO world map WGS84", xlabel="X in decimal degrees", ylabel="Y in decimal degrees")

boulder_df_geog.plot(ax=ax, markersize=52, color="springgreen")

plt.show()
```
